### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)        # x values from 0 to 10
y = x ** 2                  # y values as the square of x

# Plot
plt.plot(x, y, marker='o', color='blue')

# Add data labels (optional for clarity)
for i in range(len(x)):
    plt.text(x[i], y[i] + 2, str(y[i]), ha='center', fontsize=8)

# Customizations
plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X values')
plt.ylabel('Y = X squared')
plt.grid(True)

# Set y-ticks to only the squares
plt.yticks(y)

# Show the plot
plt.show()
```
